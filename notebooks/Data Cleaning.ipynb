Drop rows in datasets with NULL or missing values (from features dataframe as it is the only one having missing or NULL values)

```python
features = features.dropna()
features.isna().sum()
```

Merging (inner) stores and features dataframes on Store column

```python
merge_df = features.merge(stores, how = 'inner', on = 'Store')
merge_df.describe()
```

```python
#Converting date from object to datetime 

merge_df['Date'] = pd.to_datetime(merge_df['Date'])
train['Date'] = pd.to_datetime(train['Date'])
test['Date'] = pd.to_datetime(test['Date'])
```

```python
#Extracting week, year from the data and making them into new columns

merge_df['week'] = merge_df.Date.dt.isocalendar().week
merge_df['year'] = merge_df.Date.dt.isocalendar().year
```

```python
merge_df.tail()
```

Creating a new dataframe by merging (inner) the train and newly merged dataframe merge_df on store, date and isholiday columns and sorting the values by store, department and date

```python
train_df = train.merge(merge_df, how = 'inner', on = ['Store','Date','IsHoliday']).sort_values(by = ['Store','Dept','Date']).reset_index(drop = True)
train_df
```

# Data Visualisation

Scatter Plot

```python
#plot a scatter plot using the scatter function a scatter plot of weekly sales with respect to store
#plot the figure
plt.figure()

#plot the scatter plot with data from the specified column in x axis and weekly sales in y axis
plt.scatter(train_df['Store'],train_df['Weekly_Sales'])

#give y label as Weekly Sales
plt.ylabel('Weekly_Sales')

#give the x label as the column specified as parameter in the function
plt.xlabel('Store')
```

```python
#plot a scatter plot using the scatter function a scatter plot of weekly sales with respect to department
#plot the figure
plt.figure()

#plot the scatter plot with data from the specified column in x axis and weekly sales in y axis
plt.scatter(train_df['Dept'],train_df['Weekly_Sales'])

#give y label as Weekly Sales
plt.ylabel('Weekly_Sales')

#give the x label as the column specified as parameter in the function
plt.xlabel('Dept')
```

Line plots for weekly sales for each year

```python
#Filtering out mean weekly sales for the year 2010
weekly_sales_2010 = train_df[train_df['year']==2010]['Weekly_Sales'].groupby(train_df['week']).mean()

#Plot a line plot with week on x axis and sales for that particular week of the filtered year in y axis 
sns.lineplot(weekly_sales_2010.index, weekly_sales_2010.values, color = 'red')
```

```python
#Filtering out mean weekly sales for the year 2011
weekly_sales_2011 = train_df[train_df['year']==2011]['Weekly_Sales'].groupby(train_df['week']).mean()

#Plot a line plot with week on x axis and sales for that particular week of the filtered year in y axis 
sns.lineplot(weekly_sales_2011.index, weekly_sales_2011.values, color = 'black')
```

```python
#Filtering out mean weekly sales for the year 2012
weekly_sales_2012 = train_df[train_df['year']==2012]['Weekly_Sales'].groupby(train_df['week']).mean()

#Plot a line plot with week on x axis and sales for that particular week of the filtered year in y axis 
sns.lineplot(weekly_sales_2012.index, weekly_sales_2012.values, color = 'darkgreen')
```

```python
#plotting a figure
plt.figure(figsize = (20,10))

#plotting three lineplots for weekly sales for each year
sns.lineplot(weekly_sales_2010.index, weekly_sales_2010.values, color = 'crimson')
sns.lineplot(weekly_sales_2011.index, weekly_sales_2011.values, color = 'navy')
sns.lineplot(weekly_sales_2012.index, weekly_sales_2012.values, color = 'darkgreen')

plt.grid()

#plot labels and legends
plt.xticks(np.arange(1,60,step = 1))
plt.title('Average weekly sales per year', fontsize = 22)
plt.xlabel('week', fontsize = 18)
plt.ylabel('sales', fontsize = 18)
plt.legend(['2010','2011','2012'], fontsize = 18)
```
